### 1. Wprowadzanie i opis problemu

W ramach niniejszego projektu rozważany jest problem klasyfikacji binarnej, polegający na przewidywaniu poziomu dochodu osoby fizycznej na podstawie danych pochodzących ze spisu ludności. Do realizacji zadania wykorzystano klasyczny zbiór danych **Adult (Census Income)**, który stanowi standardowy punkt odniesienia w badaniach nad metodami klasyfikacyjnymi oraz regułami decyzyjnymi.

Rozpatrywany problem decyzyjny polega na **przewidywaniu, czy roczny dochód danej osoby przekracza 50 000 dolarów amerykańskich**, na podstawie zestawu jej cech demograficznych i ekonomicznych, pochodzących z danych spisowych. Formalnie jest to zadanie **klasyfikacji binarnej**, w którym każdej obserwacji przypisywana jest jedna z dwóch możliwych klas decyzyjnych:

* dochód **nieprzekraczający 50 000 USD rocznie** (`<=50K`),
* dochód **przekraczający 50 000 USD rocznie** (`>50K`).

Zmienna decyzyjna `income` stanowi etykietę klasy, natomiast pozostałe zmienne opisowe tworzą przestrzeń cech, na podstawie której konstruowany jest model klasyfikacyjny. Celem analizy jest wyznaczenie takiej reguły decyzyjnej, która na podstawie obserwowanego wektora cech maksymalizuje prawdopodobieństwo poprawnej klasyfikacji dochodu danej osoby. Problem ten ma istotne znaczenie praktyczne, ponieważ pozwala analizować zależności pomiędzy cechami demograficznymi a poziomem dochodów, a jednocześnie stanowi dobre studium przypadku do porównywania różnych metod klasyfikacji statystycznej.

Z punktu widzenia teorii statystycznych reguł decyzyjnych, zadanie to może być rozpatrywane jako problem minimalizacji ryzyka klasyfikacyjnego przy zadanej funkcji straty, co umożliwia formalną ocenę jakości zastosowanych modeli oraz porównanie ich skuteczności.

#### 3.1. Identyfikacja i analiza braków danych

Wstępna weryfikacja zbioru danych wykazała obecność niestandardowych oznaczeń braków danych (znak `?`) w zmiennych kategorialnych: `workclass`, `occupation` oraz `native-country`. Przed przystąpieniem do analizy dokonano unifikacji oznaczeń do formatu `NaN`. Poniżej przedstawiono statystyki braków oraz zbadano ich wpływ na zmienną celu, aby dobrać odpowiednią metodę imputacji.


```python
cols_wmiss = ['workclass', 'occupation', 'native-country']

# Standaryzacja braków danych (zamiana '?' na np.bnan)
df_adult = df_adult.replace('?', np.nan)

missing_stats = df_adult[cols_wmiss].isnull().sum().to_frame(name='Liczba braków')
missing_stats['Procent braków'] = (missing_stats['Liczba braków'] / len(df_adult)) * 100
print("Statystyki braków danych przed imputacją:")
print(missing_stats)

overlap = (df_adult['workclass'].isna() & df_adult['occupation'].isna()).sum()
print(f"\nLiczba rekordów z brakiem jednocześnie w 'workclass' i 'occupation': {overlap}")

# Weryfikacja nielosowości braków
mask_missing_workclass = df_adult['workclass'].isna()

pct_miss = (df_adult[mask_missing_workclass]['income'] == '>50K').mean() * 100
pct_nomiss = (df_adult[~mask_missing_workclass]['income'] == '>50K').mean() * 100

print(f"\nAnaliza wpływu braków na zmienną celu (income >50K):")
print(f" - Grupa z brakami w workclass: {pct_miss:.2f}%")
print(f" - Grupa bez braków:            {pct_nomiss:.2f}%")
print(" Wniosek: Braki są skorelowane z niższym dochodem (MNAR).")
```


**Wnioski i decyzja o imputacji**

Przeprowadzona analiza wykazała silną współzależność między brakami w zmiennych `workclass` i `occupation`. Co ważniejsze, analiza rozkładu zmiennej decyzyjnej ujawniła, że grupa osób z brakującymi danymi charakteryzuje się znacznie niższym odsetkiem osób zamożnych (>50K) w porównaniu do reszty populacji.

Sugeruje to, że braki te nie są błędem losowym, lecz niosą istotną informację semantyczną – najprawdopodobniej dotyczą osób bezrobotnych lub nigdy niepracujących. Z tego względu odrzucono metody usuwania obserwacji oraz imputacji najczęstszą wartością. Zdecydowano się na **utworzenie nowej kategorii 'Unknown'**, co pozwoli modelowi wykorzystać informację o braku danych jako cechę predykcyjną.


```python
# Implementacja decyzji: Imputacja stałą wartością 'Unknown'
for col in cols_wmiss:
    df_adult[col] = df_adult[col].fillna('Unknown')

print("Imputacja zakończona pomyślnie. Nowa liczba braków:")
print(df_adult[cols_wmiss].isnull().sum())

df_original = df_adult.copy()
```

### 5. Budowa modeli
W niniejszej sekcji przedstawiono proces budowy oraz strojenia modeli statystycznych zastosowanych do rozwiązania problemu klasyfikacji, zgodnie z założeniami projektu obejmującymi konstrukcję co najmniej trzech modeli oraz dobór ich hiperparametrów w celu uzyskania możliwie najlepszej jakości predykcji.

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from imblearn.over_sampling import SMOTE
from sklearn.metrics import confusion_matrix, auc, roc_curve, classification_report, accuracy_score, precision_score, recall_score, f1_score
import random
random.seed(42)
```

**Podział zbioru danych i przygotowanie zmiennych**

Na początku dane zostały podzielone na zmienne objaśniające (*X*) oraz zmienną docelową (*y*). Następnie cały zbiór został podzielony na część treningową i testową, przy czym 25% obserwacji przeznaczono do testowania modelu, a pozostałe 75% posłużyło do uczenia. 

Aby wszystkie zmienne numeryczne miały porównywalną skalę i nie dominowały w procesie uczenia modelu, zastosowano standaryzację. Standaryzacja polega na przekształceniu każdej wartości zmiennej tak, aby jej średnia była równa zero, a odchylenie standardowe równe jeden. Matematycznie przekształcenie można zapisać wzorem:

$$x_{\text{scaled}} = \frac{x - \mu}{\sigma}$$

gdzie:

- $x$ – oryginalna wartość zmiennej,  
- $\mu$ – średnia wartość zmiennej w zbiorze treningowym,  
- $\sigma$ – odchylenie standardowe zmiennej w zbiorze treningowym,  
- $x_{\text{scaled}}$ – przeskalowana wartość zmiennej po standaryzacji.  

Standaryzacja została wykonana **na zbiorze treningowym**, a te same parametry ($\mu$ i $\sigma$) zastosowano do danych testowych, co zapewnia, że model nie „widzi” danych testowych podczas uczenia.

```python
# podział zbioru na treningowy i testowy 
X = df_adult.drop(columns=['income_binary'])
y = df_adult['income_binary']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# standaryzacja zmiennych numerycznych (fit na train, transform na train i test)
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])
```

**Nierównowaga klas**

W danych możnna zaobserwować znaczną nierównowagę liczebności klas zmiennej celu: większość osób zarabia poniżej 50 000 USD rocznie, natomiast osoby z wyższymi dochodami stanowią znacznie mniejszość. Taka nierównowaga może prowadzić do tego, że model będzie faworyzował klasę dominującą, a słabo rozpoznawał klasę mniejszościową.

Aby złagodzić ten problem, zastosowano technikę **SMOTE (Synthetic Minority Over-sampling Technique)**, która polega na sztucznym powiększeniu liczby obserwacji klasy mniejszościowej poprzez generowanie nowych przykładów w oparciu o istniejące dane. Dzięki temu zbiór treningowy staje się bardziej zrównoważony, co zwiększa szansę modelu na poprawne rozpoznanie rzadziej występującej klasy.

![SMOTE](https://www.researchgate.net/profile/Joao-Tavares-23/publication/347937180/figure/fig3/AS:973429209563136@1609095017080/llustration-of-the-SMOTE-oversampling-approach.ppm)

```python
smote = SMOTE(random_state=42);
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train);
```

W ramach projektu przetestowane zostanie też drugie podejścia do radzenia sobie z nierównowagą klas: **Class weight**, które polega na przypisaniu większej wagi klasie mniejszościowej bez zmiany liczby obserwacji, dzięki czemu model automatycznie „zwraca większą uwagę” na rzadziej występujące przypadki.

Porównanie obu podejść pozwoli ocenić, która metoda lepiej sprawdza się w kontekście klasyfikacji dochodów na podstawie danych z badania spisowego.

**Funkcje pomocnicze**

W ramach oceny jakości modelu, stworzone zostały 3 funkcje pomocnicze:

1. **`plot_confusion_matrices`**  
   Funkcja wyświetla dwie macierze pomyłek obok siebie – dla zbioru treningowego i testowego. Pozwala szybko ocenić, ile obserwacji każdej klasy zostało poprawnie lub błędnie sklasyfikowanych przez model.

2. **`plot_roc_auc_curve`**  
   Funkcja rysuje krzywe ROC dla danych treningowych i testowych wraz z wartościami AUC. Krzywa ROC ilustruje zależność między odsetkiem fałszywych alarmów a odsetkiem prawidłowo wykrytych przypadków, a AUC mierzy ogólną zdolność modelu do rozróżniania klas.

3. **`print_report`**  
   Funkcja drukuje klasyczny raport klasyfikacyjny (`precision`, `recall`, `f1-score`) dla zbiorów treningowego i testowego. Pozwala porównać jakość predykcji modelu na danych uczących i nowych, nieznanych danych.


```python
def plot_confusion_matrices(y_train, y_train_hat, y_test, y_test_hat):
    cm_train, cm_test = confusion_matrix(y_train, y_train_hat), confusion_matrix(y_test, y_test_hat)

    plt.figure(figsize=(10, 4))

    plt.subplot(1,2,1)
    sns.heatmap(cm_train, annot=True, fmt="d", cmap="Blues", 
                xticklabels=["Predicted 0", "Predicted 1"], 
                yticklabels=["Actual 0", "Actual 1"])
    plt.title("Train Confusion Matrix")
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')

    plt.subplot(1,2,2)
    sns.heatmap(cm_test, annot=True, fmt="d", cmap="Blues", 
                xticklabels=["Predicted 0", "Predicted 1"], 
                yticklabels=["Actual 0", "Actual 1"])
    plt.title("Test Confusion Matrix")
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')

    plt.show()


def plot_roc_auc_curve(X_train, y_train, X_test, y_test, model):
    y_train_lr_proba = model.predict_proba(X_train)[:,1]
    fprt, tprt, _ = roc_curve(y_train, y_train_lr_proba)
    auc_roct = auc(fprt, tprt)

    y_test_lr_proba = model.predict_proba(X_test)[:,1]
    fprv, tprv, _ = roc_curve(y_test, y_test_lr_proba)
    auc_rocv = auc(fprv, tprv)

    plt.figure(figsize=(5, 4))

    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC+AUC")

    plt.plot([0, 1], [0, 1], color="grey", linestyle="--", label="Random, AUC = 0.5")
    plt.plot([0, 0], [0, 1], color="navy", linestyle=":", label="Wizard, AUC = 1.0")
    plt.plot([0, 1], [1, 1], color="navy", linestyle=":")

    plt.plot(fprt, tprt, color="orange", label="Model - train, AUC = %0.2f" % auc_roct)
    plt.plot(fprv, tprv, color="red", label="Model - val, AUC = %0.2f" % auc_rocv)
    plt.legend(loc="lower right");

    plt.show()


def print_report(y_train, y_train_hat, y_test, y_test_hat, target_names):
    print("Training data:\n", classification_report(y_train, y_train_hat, target_names=target_names), "\n\nTest data:\n", classification_report(y_test, y_test_hat, target_names=target_names))
```

#### 5.1. Regresja logistyczna

Regresja logistyczna jest metodą statystyczną służącą do **modelowania prawdopodobieństwa przynależności obserwacji do jednej z dwóch klas**. W przeciwieństwie do regresji liniowej, która przewiduje wartości ciągłe, regresja logistyczna stosuje funkcję logistyczną (sigmoidę), dzięki czemu wyjście modelu mieści się w przedziale [0,1] i może być interpretowane jako prawdopodobieństwo.

**Model matematyczny**

Regresję logistyczną można opisać równaniem:

$$P(Y=1 \mid X) = \sigma(z) = \frac{1}{1 + e^{-z}}$$

gdzie:

$$z = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_p x_p$$

- $P(Y=1 \mid X)$ – prawdopodobieństwo, że obserwacja należy do klasy pozytywnej (1),  
- $X = (x_1, x_2, \dots, x_p)$ – wektor zmiennych objaśniających,  
- $\beta_0$ – wyraz wolny (intercept),  
- $\beta_1, \dots, \beta_p$ – współczynniki regresji logistycznej,  
- $\sigma(z)$ – funkcja sigmoidalna, przekształcająca wartość liniową $z$ w przedział $[0,1]$.

**Funkcja kosztu**

W regresji logistycznej stosuje się **log-loss (cross-entropy)** jako funkcję kosztu, którą minimalizuje model:

$$\mathcal{L}(\beta) = - \frac{1}{n} \sum_{i=1}^{n} \Big[ y_i \log(\hat{y}_i) + (1-y_i) \log(1-\hat{y}_i) \Big]$$

gdzie:

- $n$ – liczba obserwacji,  
- $y_i \in \{0,1\}$ – rzeczywista klasa $i$-tej obserwacji,  
- $\hat{y}_i = P(Y=1 \mid X_i)$ – przewidywane prawdopodobieństwo przynależności do klasy 1.  

Minimalizacja tej funkcji pozwala znaleźć takie wartości współczynników $\beta$, które najlepiej dopasowują model do danych treningowych.

**Predykcja klasy**

Po wytrenowaniu modelu, klasy przypisuje się zazwyczaj na podstawie progu 0.5:

$$
\hat{Y} =
\begin{cases} 
1 & \text{jeżeli } P(Y=1 \mid X) \geq 0.5 \\
0 & \text{jeżeli } P(Y=1 \mid X) < 0.5
\end{cases}
$$

Regresja logistyczna jest prostym, interpretowalnym modelem, który dobrze sprawdza się przy problemach klasyfikacji binarnej, a jej współczynniki można interpretować jako wpływ danej cechy na logarytm szans ($\log$ odds) zajścia zdarzenia.


```python
from sklearn.linear_model import LogisticRegression
```

Model ten pozwala ocenić skuteczność predykcji na najprostszej formie regresji logitowej, a jego wyniki stanowią punkt odniesienia przy porównywaniu innych metod i wariantów przetwarzania danych.

##### 5.1.2. Dobór parametrów do modelu

W celu optymalizacji jakości predykcji podstawowego modelu regresji logistycznej przeprowadzono **dobór hiperparametrów** przy użyciu metody GridSearchCV. Przebadano trzy główne parametry: solver, współczynnik regularizacji $C$ oraz typ kary (*penalty*).

**Solver**

Wybrano solver **`saga`**, który jest odpowiedni dla dużych zbiorów danych oraz obsługuje wszystkie typy kar: L1, L2 i elasticnet. Ponadto `saga` dobrze radzi sobie ze zmiennymi binarnymi i one-hot encodingiem, co odpowiada strukturze danego zbioru, w którym większość zmiennych kategorialnych została zakodowana w postaci binarnej.

**Współczynnik regularizacji $C$**

Parametr $C$ odwrotnie kontroluje siłę regularyzacji modelu – mniejsze wartości $C$ zwiększają regularyzację, co może zmniejszyć ryzyko przeuczenia, natomiast większe wartości pozwalają modelowi lepiej dopasować się do danych treningowych.  

Dla eksperymentu przyjęto zakres wartości: $C \in [0.001, 0.01, 0.046, 0.215, 1, 4.64, 10]$.

Wartości te zostały dobrane logarytmicznie, aby umożliwić testowanie zarówno silnej regularyzacji, jak i prawie całkowitego jej braku. Taki zakres pozwala zidentyfikować optymalny kompromis między dopasowaniem modelu a jego generalizacją.

**Typ kary (*penalty*)**

Rozważano trzy rodzaje kar:

- **L1** – wymusza rzadkość współczynników, co może prowadzić do automatycznej selekcji cech,  
- **L2** – klasyczna regularyzacja, która równomiernie zmniejsza wartości wszystkich współczynników,  
- **ElasticNet** – kombinacja L1 i L2, pozwalająca jednocześnie na rzadkość współczynników i stabilizację modelu.  

Nie rozważano opcji `None`, ponieważ wybrany solver `saga` wymaga określenia rodzaju regularyzacji. Brak regularyzacji w dużych zbiorach danych mógłby dodatkowo prowadzić do przeuczenia i niestabilnych estymacji współczynników.

**Procedura GridSearchCV**

Parametry zostały przebadane w ramach **5-krotnej walidacji krzyżowej** na zbiorze treningowym. Jako kryterium optymalizacji wybrano dokładność (*accuracy*), a testowane kombinacje parametrów pozwoliły zidentyfikować konfigurację zapewniającą najlepszą równowagę między dokładnością na zbiorze treningowym a zdolnością do generalizacji na danych testowych.


```python
lr_basic = LogisticRegression(solver="saga", class_weight='balanced', max_iter=1000, random_state=42, l1_ratio=0.5)

C_grid = np.logspace(-3, 1, 7) # np.linspace(0.01, 10, 100)
penalty = ["l1", "l2", "elasticnet"]
param_grid = {"penalty": penalty, "C": C_grid}
```

```python
grid_search = GridSearchCV(
    estimator=lr_basic,
    param_grid=param_grid,
    scoring="accuracy",
    cv=5,
    n_jobs=-1
)

grid_search.fit(X_train, y_train)
```

```python
print("GridSearchCV best parameters")
print(f"C:\t\t{grid_search.best_params_['C']}")
print(f"penalty:\t{grid_search.best_params_['penalty']}")

print("\nBest scores")
print(f"GridSearch:\t{grid_search.best_score_:.10f}")
```

Najlepszy model wybrany w wyniku GridSearchCV osiągnął średnią dokładność na walidacji krzyżowej równą **0.81**.

Optymalne parametry modelu to: **C = $0.464$** oraz **penalty = L1**. Model ten został przyjęty jako konfiguracja referencyjna do dalszej oceny na danych testowych.

```python
results = pd.DataFrame(grid_search.cv_results_)

pivot = results.pivot_table(
    values="mean_test_score",
    index="param_C",
    columns="param_penalty"
)

pivot = pivot.sort_index()

# heatmapa
plt.figure(figsize=(8, 6))
plt.imshow(pivot, aspect="auto")
plt.colorbar(label="Średnie accuracy")

plt.xticks(range(len(pivot.columns)), pivot.columns)
plt.yticks(range(len(pivot.index)), np.round(pivot.index, 4))

plt.xlabel("Funkcja kary")
plt.ylabel("Wartość C")
plt.title("Regresja logistyczna - wyniki GridSearchCV")

for i in range(pivot.shape[0]):
    for j in range(pivot.shape[1]):
        value = pivot.iloc[i, j]
        plt.text(
            j, i, f"{value:.3f}",
            ha="center", va="center",
            color="black", fontsize=9
        )

plt.tight_layout()
plt.show()
```

Analiza wyników GridSearchCV pokazuje, że dokładność modelu rośnie wraz ze wzrostem wartości parametru regularyzacji $C$ od bardzo małych wartości (0.001) do ok. 0.464, po czym stabilizuje się dla większych $C$. Najlepszy wynik uzyskano dla **C = 0.464** i **penalty = L1**, co sugeruje, że umiarkowana regularyzacja z karą L1 zapewnia optymalny kompromis między dopasowaniem modelu do danych treningowych a jego zdolnością do generalizacji.  

Ponadto można zauważyć, że dla wszystkich testowanych wartości $C$ różnice między typami kary są niewielkie, co wskazuje na stabilność modelu względem wyboru penalty.

#### 5.3. XGBoost

XGBoost (Extreme Gradient Boosting) to zaawansowany algorytm oparty na metodzie wzmacniania gradientowego (gradient boosting). Buduje sekwencyjnie drzewa decyzyjne, gdzie każde kolejne drzewo koryguje błędy poprzednich. XGBoost jest znany z wysokiej skuteczności w zadaniach klasyfikacji tabelarycznej oraz wbudowanej obsługi niezbalansowania klas poprzez parametr `scale_pos_weight`.

```python
from xgboost import XGBClassifier

# Obliczenie wagi dla klasy mniejszościowej (obsługa niezbalansowania)
scale_pos_weight = len(y_train[y_train == 0]) / len(y_train[y_train == 1])
print(f"scale_pos_weight: {scale_pos_weight:.2f}")
```

##### 5.3.1. Model bazowy XGBoost

```python
# Model bazowy XGBoost
xgb_baseline = XGBClassifier(
    scale_pos_weight=scale_pos_weight,
    random_state=42,
    eval_metric='logloss'
)

xgb_baseline.fit(X_train, y_train)

# Predykcje
y_train_xgb_base = xgb_baseline.predict(X_train)
y_test_xgb_base = xgb_baseline.predict(X_test)
```

```python
# Ewaluacja modelu bazowego
plot_confusion_matrices(y_train, y_train_xgb_base, y_test, y_test_xgb_base)
plot_roc_auc_curve(X_train, y_train, X_test, y_test, xgb_baseline)
print_report(y_train, y_train_xgb_base, y_test, y_test_xgb_base, target_names)
```

##### 5.3.2. Tuning hiperparametrów

Do optymalizacji hiperparametrów wykorzystano GridSearchCV z walidacją krzyżową (5-fold). Przeszukiwane parametry:
- **n_estimators**: liczba drzew (50, 100, 200)
- **max_depth**: maksymalna głębokość drzewa (3, 5, 7)
- **learning_rate**: współczynnik uczenia (0.01, 0.1, 0.2)
- **subsample**: frakcja próbek do budowy każdego drzewa (0.8, 1.0)
- **colsample_bytree**: frakcja cech do budowy każdego drzewa (0.8, 1.0)

```python
# Definicja siatki hiperparametrów
xgb_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

# Model bazowy do tuningu
xgb_base = XGBClassifier(
    scale_pos_weight=scale_pos_weight,
    random_state=42,
    eval_metric='logloss'
)

# GridSearchCV
xgb_grid_search = GridSearchCV(
    estimator=xgb_base,
    param_grid=xgb_param_grid,
    scoring='f1',  # F1-score ze względu na niezbalansowanie klas
    cv=5,
    n_jobs=-1,
    verbose=1
)

print("Rozpoczęcie GridSearchCV dla XGBoost...")
print(f"Liczba kombinacji: {3*3*3*2*2} = 108")
xgb_grid_search.fit(X_train, y_train)
```

```python
# Wyświetlenie najlepszych parametrów
print("=" * 50)
print("NAJLEPSZE HIPERPARAMETRY XGBoost:")
print("=" * 50)
for param, value in xgb_grid_search.best_params_.items():
    print(f"  {param}: {value}")
print(f"\nNajlepszy wynik F1 (CV): {xgb_grid_search.best_score_:.4f}")
```

##### 5.3.3. Ewaluacja najlepszego modelu XGBoost

```python
# Najlepszy model
xgb_best = xgb_grid_search.best_estimator_

# Predykcje
y_train_xgb_best = xgb_best.predict(X_train)
y_test_xgb_best = xgb_best.predict(X_test)

# Ewaluacja
plot_confusion_matrices(y_train, y_train_xgb_best, y_test, y_test_xgb_best)
plot_roc_auc_curve(X_train, y_train, X_test, y_test, xgb_best)
print_report(y_train, y_train_xgb_best, y_test, y_test_xgb_best, target_names)
```

##### 5.3.4. Ważność cech (Feature Importance)

XGBoost umożliwia analizę ważności cech, co pozwala zidentyfikować zmienne o największym wpływie na predykcję dochodu.

```python
# Ważność cech
feature_importance = pd.DataFrame({
    'feature': X_train.columns,
    'importance': xgb_best.feature_importances_
}).sort_values('importance', ascending=False)

# Wykres top 15 najważniejszych cech
plt.figure(figsize=(10, 8))
top_features = feature_importance.head(15)
plt.barh(range(len(top_features)), top_features['importance'].values, color='steelblue', edgecolor='black')
plt.yticks(range(len(top_features)), top_features['feature'].values)
plt.xlabel('Ważność (Importance)')
plt.title('Top 15 najważniejszych cech - XGBoost')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

print("\nTop 10 najważniejszych cech:")
print(feature_importance.head(10).to_string(index=False))
```

**Podsumowanie modelu XGBoost:**

Model XGBoost po tuningu hiperparametrów wykazuje wysoką skuteczność w klasyfikacji dochodu. Analiza ważności cech potwierdza wcześniejsze obserwacje z EDA - zmienne takie jak `marital-status`, `education-num`, `age` oraz `capital-gain` mają największy wpływ na predykcję. Model dobrze radzi sobie z niezbalansowaniem klas dzięki zastosowaniu parametru `scale_pos_weight`.

- Regresja logistyczna jest prostym i stabilnym modelem, który dobrze radzi sobie z niezbalansowaniem klas, choć osiąga nieco niższą dokładność ogólną.

#### 6.1. Analiza porównawcza modeli klasyfikacyjnych

**Zdolność separacji i stabilność (ROC AUC & Accuracy)**

Wszystkie modele osiągnęły wysokie wyniki ROC AUC na zbiorze testowym (0.9–0.93), co świadczy o ich dużej skuteczności w rozróżnianiu klas niezależnie od progu odcięcia.
* **RandomForest** osiągnął najwyższe **Accuracy (0.87)** na zbiorze testowym.
* **XGBoost** wykazuje najwyższe testowe **ROC AUC (0.93)**, co sugeruje, że potencjalnie może on działać jeszcze lepiej po precyzyjnym dostrojeniu progu klasyfikacji.
* **Regresja Logistyczna** (Accuracy 0.81) wypada najsłabiej pod kątem ogólnym, ale jej wyniki są identyczne dla zbioru treningowego i testowego, co czyni ją modelem wzorcowo stabilnym.

**Charakterystyka predykcji klasy mniejszościowej (>50K)**
* **RandomForest:** Posiada najwyższą precyzję dla klasy >50K (**0.80** na teście).
* **Interpretacja:** Jeśli RF przewiduje wysoki dochód, mamy 80% pewności, że tak jest. Jednak model ten "boi się" ryzykować, przez co pomija wielu zamożnych podatników (Recall tylko **0.59**). Potwierdza to macierz błędów: aż 1218 błędów typu *False Negative*.
* **XGBoost:** Osiąga najwyższy Recall dla klasy >50K (**0.86** na teście).
* **Interpretacja:** XGBoost jest najskuteczniejszy w wyłapywaniu osób zamożnych (tylko 421 pominiętych przypadków w macierzy testowej). Odbywa się to jednak kosztem większej liczby błędów typu *False Positive* (1635 osób błędnie uznanych za zamożne).
* **Regresja Logistyczna:** Mimo prostej budowy, oferuje bardzo solidny Recall (**0.84**). Jest to model "bezpieczny", który dobrze radzi sobie z nierównowagą klas bez skomplikowanych struktur drzewiastych.

**Analiza macierzy błędów (Confusion Matrix)**
Macierze błędów wizualizują to, co widać w metrykach:
* **RandomForest** dominuje w poprawnym klasyfikowaniu klasy 0 (8822 trafienia), ale jest zbyt ostrożny w klasie 1.
* **XGBoost** i **Regresja Logistyczna** znacznie lepiej radzą sobie z przekątną dla klasy 1, co sugeruje ich lepszą użyteczność w zadaniach, gdzie priorytetem jest znalezienie jak największej liczby osób z grupy >50K.

**Wniosek końcowy:**  
Wybór odpowiedniego modelu zależy od przyjętych priorytetów analitycznych:  
* **XGBoost** wykazuje przewagę w scenariuszach, gdzie kluczowa jest maksymalna wykrywalność klasy mniejszościowej (wysoki Recall) i dopuszczalna jest większa liczba błędnych wskazań dodatnich.
