### 1. Wprowadzanie i opis problemu

W ramach niniejszego projektu rozważany jest problem klasyfikacji binarnej, polegający na przewidywaniu poziomu dochodu osoby fizycznej na podstawie danych pochodzących ze spisu ludności. Do realizacji zadania wykorzystano klasyczny zbiór danych **Adult (Census Income)**, który stanowi standardowy punkt odniesienia w badaniach nad metodami klasyfikacyjnymi oraz regułami decyzyjnymi.

Rozpatrywany problem decyzyjny polega na **przewidywaniu, czy roczny dochód danej osoby przekracza 50 000 dolarów amerykańskich**, na podstawie zestawu jej cech demograficznych i ekonomicznych, pochodzących z danych spisowych. Formalnie jest to zadanie **klasyfikacji binarnej**, w którym każdej obserwacji przypisywana jest jedna z dwóch możliwych klas decyzyjnych:

* dochód **nieprzekraczający 50 000 USD rocznie** (`<=50K`),
* dochód **przekraczający 50 000 USD rocznie** (`>50K`).

Zmienna decyzyjna `income` stanowi etykietę klasy, natomiast pozostałe zmienne opisowe tworzą przestrzeń cech, na podstawie której konstruowany jest model klasyfikacyjny. Celem analizy jest wyznaczenie takiej reguły decyzyjnej, która na podstawie obserwowanego wektora cech maksymalizuje prawdopodobieństwo poprawnej klasyfikacji dochodu danej osoby. Problem ten ma istotne znaczenie praktyczne, ponieważ pozwala analizować zależności pomiędzy cechami demograficznymi a poziomem dochodów, a jednocześnie stanowi dobre studium przypadku do porównywania różnych metod klasyfikacji statystycznej.

Z punktu widzenia teorii statystycznych reguł decyzyjnych, zadanie to może być rozpatrywane jako problem minimalizacji ryzyka klasyfikacyjnego przy zadanej funkcji straty, co umożliwia formalną ocenę jakości zastosowanych modeli oraz porównanie ich skuteczności.

#### 3.1. Identyfikacja i analiza braków danych

Wstępna weryfikacja zbioru danych wykazała obecność niestandardowych oznaczeń braków danych (znak `?`) w zmiennych kategorycznych: `workclass`, `occupation` oraz `native-country`. Przed przystąpieniem do analizy dokonano unifikacji oznaczeń do formatu `NaN`. Poniżej przedstawiono statystyki braków oraz zbadano ich wpływ na zmienną celu, aby dobrać odpowiednią metodę imputacji.


```python
cols_wmiss = ['workclass', 'occupation', 'native-country']

# Standaryzacja braków danych (zamiana '?' na np.bnan)
df_adult = df_adult.replace('?', np.nan)

missing_stats = df_adult[cols_wmiss].isnull().sum().to_frame(name='Liczba braków')
missing_stats['Procent braków'] = (missing_stats['Liczba braków'] / len(df_adult)) * 100
print("Statystyki braków danych przed imputacją:")
print(missing_stats)

overlap = (df_adult['workclass'].isna() & df_adult['occupation'].isna()).sum()
print(f"\nLiczba rekordów z brakiem jednocześnie w 'workclass' i 'occupation': {overlap}")

# Weryfikacja nielosowości braków
mask_missing_workclass = df_adult['workclass'].isna()

pct_miss = (df_adult[mask_missing_workclass]['income'] == '>50K').mean() * 100
pct_nomiss = (df_adult[~mask_missing_workclass]['income'] == '>50K').mean() * 100

print(f"\nAnaliza wpływu braków na zmienną celu (income >50K):")
print(f" - Grupa z brakami w workclass: {pct_miss:.2f}%")
print(f" - Grupa bez braków:            {pct_nomiss:.2f}%")
print(" Wniosek: Braki są skorelowane z niższym dochodem (MNAR).")
```


**Wnioski i decyzja o imputacji**

Przeprowadzona analiza wykazała silną współzależność między brakami w zmiennych `workclass` i `occupation`. Co ważniejsze, analiza rozkładu zmiennej decyzyjnej ujawniła, że grupa osób z brakującymi danymi charakteryzuje się znacznie niższym odsetkiem osób zamożnych (>50K) w porównaniu do reszty populacji.

Sugeruje to, że braki te nie są błędem losowym, lecz niosą istotną informację semantyczną – najprawdopodobniej dotyczą osób bezrobotnych lub nigdy niepracujących. Z tego względu odrzucono metody usuwania obserwacji oraz imputacji najczęstszą wartością. Zdecydowano się na **utworzenie nowej kategorii 'Unknown'**, co pozwoli modelowi wykorzystać informację o braku danych jako cechę predykcyjną.


```python
# Implementacja decyzji: Imputacja stałą wartością 'Unknown'
for col in cols_wmiss:
    df_adult[col] = df_adult[col].fillna('Unknown')

print("Imputacja zakończona pomyślnie. Nowa liczba braków:")
print(df_adult[cols_wmiss].isnull().sum())

df_checkpoint = df_adult.copy()
print(df_checkpoint)
```

#### 4.1. Rozkłady zmiennych numerycznych
W pierwszej kolejności przeanalizowano rozkłady zmiennych ciągłych i porządkowych. Z analizy wyłączono zmienne binarne (posiadające tylko 2 wartości), skupiając się na cechach o większej różnorodności wartości.

```python
skip_cols = ['income']

for col in df_adult.columns:
    if col in skip_cols:
        continue
    
    if len(df_adult[col].unique()) > 2:
        plt.figure(figsize=(6, 4))
        
        plt.hist(df_adult[col], bins=30, color='skyblue', edgecolor='black', alpha=0.7)
        
        plt.title(f'Rozkład zmiennej: {col}')
        plt.xlabel(col)
        plt.ylabel('Liczebność obserwacji')
        plt.grid(axis='y', alpha=0.5)
        plt.show()
```

**Wnioski z analizy rozkładów zmiennych:**

1.  **`age` (Wiek):**
    Rozkład jest wyraźnie **prawoskośny**. Najliczniejszą grupę stanowią osoby w wieku produkcyjnym (20–45 lat). Wraz z wiekiem liczebność grupy spada łagodnie, co jest naturalnym odzwierciedleniem struktury siły roboczej.

2.  **`fnlwgt` (Waga demograficzna):**
    Rozkład jest zbliżony do log-normalnego, bez widocznych anomalii. Jest to zmienna techniczna (waga statystyczna przypisana przez Census Bureau) i jej rozkład nie niesie bezpośredniej interpretacji ekonomicznej dotyczącej zarobków jednostki.

3.  **`education-num` (Lata edukacji):**
    Zmienna wykazuje charakter **wielomodalny** (posiada kilka szczytów). Wyraźne "piki" odpowiadają momentom kończenia konkretnych etapów edukacji w systemie USA:

    * **10:** Częściowe studia (Some-college).
    * **13:** Licencjat (Bachelors).
    Potwierdza to skokowy charakter zdobywania kwalifikacji.


4.  **`capital-gain` (Zyski kapitałowe):**
    Zmienna o **ekstremalnej skośności**. Olbrzymi słupek przy wartości **0** oznacza, że przytłaczająca większość populacji nie posiada dochodów z inwestycji giełdowych. Na prawym krańcu wykresu widoczne jest jednak małe skupisko wartości bardzo wysokich (tzw. outliery), co sugeruje obecność wąskiej grupy zamożnych inwestorów.

5.  **`capital-loss` (Straty kapitałowe):**
    Rozkład analogiczny do zysków kapitałowych – zdecydowana dominacja wartości **0**. Zmienna ta jest rzadka (sparse), co oznacza, że niesie informację tylko dla niewielkiego ułamka obserwacji.

#### 4.2. Analiza relacji: Zmienne numeryczne vs Zmienna celu (Income)
W tej sekcji zbadano, czy rozkłady kluczowych zmiennych numerycznych różnią się pomiędzy grupami dochodowymi. Wykorzystano wykresy pudełkowe, które pozwalają porównać mediany (pozioma kreska w pudełku) oraz rozstęp międzykwartylowy dla klasy `<=50K` (0) oraz `>50K` (1).

```python
target_vars = ['age', 'education-num', 'hours-per-week']

plt.figure(figsize=(16, 6))

for i, col in enumerate(target_vars):
    plt.subplot(1, 3, i+1)
    sns.boxplot(x='income_binary', y=col, data=df_adult, palette='Set1')
    plt.title(f'{col} wg grupy dochodowej')
    plt.xticks([0, 1], ['<=50K', '>50K'])
    plt.xlabel('')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

**Wnioski z analizy relacji zmiennych:**

Wykresy pudełkowe ujawniają kluczowe różnice w profilu statystycznym osób o niższych (<=50K) i wyższych (>50K) dochodach:

1. **`age` (Wiek):**
    - **Przesunięcie mediany:** Widzimy wyraźny skok. Mediana wieku dla grupy biedniejszej (czerwone pudełko) wynosi około **34 lat**, podczas gdy dla grupy zamożnej (niebieskie pudełko) to około **44-45 lat**.
    - **Wniosek:** Dochód jest silnie skorelowany z wiekiem. Sugeruje to, że wyższe zarobki przychodzą wraz z doświadczeniem zawodowym i stażem pracy. Model z pewnością wykorzysta wiek jako ważną cechę.

2. **`education-num` (Lata edukacji):**
    - **Najsilniejszy sygnał:** To zmienna najlepiej separująca obie grupy.
    - **Różnica poziomów:** Mediana dla grupy `<=50K` to **9** (ukończona szkoła średnia). Dla grupy `>50K` mediana skacze aż do **13** (Licencjat/Bachelors).
    - **Rozkład:** Górna krawędź niebieskiego pudełka sięga 16 (Doktorat), co pokazuje, że w grupie zamożnej wyższe wykształcenie jest standardem, a nie wyjątkiem.

4. **`hours-per-week` (Godziny pracy):**
    - **Standard vs Nadgodziny:** Dla grupy uboższej mediana wynosi sztywno **40 godzin** (standardowy etat), a dolny wąs sięga nisko (praca dorywcza).
    - **Pracowitość popłaca:** W grupie zamożnej mediana jest wyższa (ok. **45-50 godzin**), a całe "pudełko" jest przesunięte w górę. Oznacza to, że osoby zarabiające dużo, statystycznie częściej pracują w nadgodzinach lub na wymagających stanowiskach kierowniczych.

**Podsumowanie:**
Zmienne `education-num`, `age` oraz `hours-per-week` wykazują silną zdolność dyskryminacyjną (rozróżniającą) i będą kluczowymi predyktorami w budowanym modelu klasyfikacji.

#### 4.3. Analiza korelacji zmiennych numerycznych

```python
cols_to_check = ['age', 'education-num', 'hours-per-week', 'capital-gain', 'capital-loss', 'income_binary']

plt.figure(figsize=(10, 8))

corr_matrix = df_adult[cols_to_check].corr(method='spearman')

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, vmin=-1, vmax=1)
plt.title('Macierz korelacji (zmienne numeryczne)')
plt.show()
```

**Kluczowe determinanty dochodu:**
- Najsilniejszą dodatnią korelację ze zmienną celu wykazują:
  - **`education-num` (0.33):** Potwierdza to wcześniejszą obserwację z wykresów pudełkowych – poziom wykształcenia jest najsilniejszym pojedynczym predyktorem wysokich zarobków w tym zbiorze.
  - **`capital-gain` (0.28):** Mimo że zmienna ta jest rzadka (dużo zer), jej wystąpienie jest silnie skorelowane z bogactwem.
  - **`age` (0.27) oraz `hours-per-week` (0.27):** Wiek i zaangażowanie czasowe w pracę mają niemal identyczną, istotną siłę wpływu na dochód.

**Brak silnej współliniowości (Multicollinearity):**
- Analizując korelacje *pomiędzy* samymi zmiennymi niezależnymi (np. `age` vs `education-num`), nie obserwujemy żadnych wysokich wartości (wszystkie są poniżej 0.20).
- Jest to **bardzo dobra informacja** w kontekście modelowania:
  - Oznacza to, że każda z tych zmiennych wnosi do modelu unikalną, niezależną informację.
  - Nie ma potrzeby usuwania zmiennych z powodu ich wzajemnej redundancji.

**Wnioski końcowe z EDA:**
- Zbiór danych posiada silne sygnały predykcyjne (zarówno w zmiennych demograficznych, jak i finansowych) i jest dobrze przygotowany do etapu modelowania.
- Zmienne takie jak `fnlwgt` (brak korelacji) zostały słusznie zidentyfikowane jako szum.
- Zestaw `education`, `age`, `hours`, `gain` stanowi solidny fundament dla algorytmów klasyfikacji.

```python
df_org = df_checkpoint.copy()
df_org["income"] = y

df_org["income"] = df_org["income"].apply(lambda x: 1 if x in ['>50K', '>50K.'] else 0)

plt.figure(figsize=(14, 10))

cat_vars = ['marital-status', 'relationship', 'sex', 'workclass', 'race', 'occupation']

for i, col in enumerate(cat_vars):
    plt.subplot(2, 3, i+1)
    
    # Seaborn sam ogarnie kolory i słupki dla każdej kategorii tekstowej!
    sns.barplot(x=col, y='income', data=df_org, errorbar=None)
    
    plt.title(f'Szansa na zarobki >50K wg: {col}')
    plt.ylabel('Prawdopodobieństwo')
    plt.xlabel('')
    plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

### 5. Budowa modeli
W niniejszej sekcji przedstawiono proces budowy oraz strojenia modeli statystycznych zastosowanych do rozwiązania problemu klasyfikacji, zgodnie z założeniami projektu obejmującymi konstrukcję co najmniej trzech modeli oraz dobór ich hiperparametrów w celu uzyskania możliwie najlepszej jakości predykcji.

```python
from sklearn.model_selection import train_test_split

from sklearn.metrics import confusion_matrix, auc, roc_curve, classification_report, accuracy_score, precision_score, recall_score, f1_score

import random
random.seed(42)
```

```python
# podział zbioru na treningowy i testowy 
X = df_adult.drop(columns=['income_binary'])
y = df_adult['income_binary']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=16)
```

#### 5.1. Regresja logistyczna

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(random_state=42)
```

```python
logreg.fit(X_train, y_train)
```

```python
y_train_hat = logreg.predict(X_train)
y_test_hat = logreg.predict(X_test)
```

```python
cm_train = confusion_matrix(y_train, y_train_hat)
cm_test = confusion_matrix(y_test, y_test_hat)

def plot_confusion_matrix(cm, title, place):
    plt.subplot(1,2,place)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", 
                xticklabels=["Predicted 0", "Predicted 1"], 
                yticklabels=["Actual 0", "Actual 1"])
    plt.title(title)
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')

plt.figure(figsize=(15, 5))
plot_confusion_matrix(cm_train, "Train Confusion Matrix", 1)
plot_confusion_matrix(cm_test, "Test Confusion Matrix", 2)
plt.show()
```

```python
y_train_proba = logreg.predict_proba(X_train)[:,1]
fprt, tprt, _ = roc_curve(y_train, y_train_proba)
auc_roct = auc(fprt, tprt)

y_test_proba = logreg.predict_proba(X_test)[:,1]
fprv, tprv, _ = roc_curve(y_test, y_test_proba)
auc_rocv = auc(fprv, tprv)

plt.figure()

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC+AUC")

plt.plot([0, 1], [0, 1], color="grey", linestyle="--", label="Random, AUC = 0.5")
plt.plot([0, 0], [0, 1], color="navy", linestyle=":", label="Wizard, AUC = 1.0")
plt.plot([0, 1], [1, 1], color="navy", linestyle=":")

plt.plot(fprt, tprt, color="orange", label="Model - train, AUC = %0.2f" % auc_roct)
plt.plot(fprv, tprv, color="red", label="Model - val, AUC = %0.2f" % auc_rocv)
plt.legend(loc="lower right");
```

```python
print("Training data:\n", classification_report(y_train, y_train_hat), "\n\nTest data:\n", classification_report(y_test, y_test_hat))
```

```python
print("Training data:")
print(f"Precission\t{precision_score(y_train, y_train_hat)}\nRecall\t\t{recall_score(y_train, y_train_hat)}\nF1\t\t{f1_score(y_train, y_train_hat)}\nAccuracy\t{accuracy_score(y_train, y_train_hat)}")

print("\nTest data:")
print(f"Precission\t{precision_score(y_test, y_test_hat)}\nRecall\t\t{recall_score(y_test, y_test_hat)}\nF1\t\t{f1_score(y_test, y_test_hat)}\nAccuracy\t{accuracy_score(y_test, y_test_hat)}")
```
